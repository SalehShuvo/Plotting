## Double Axes marking

```python
import numpy as np
import matplotlib.pyplot as plt

import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.serif'] = ['DejaVu Serif']



# Sample data for plotting
np.random.seed(0)
x = np.arange(1, 6)
data = [np.random.normal(loc=i, scale=0.3, size=100) for i in x]
mean_values = [np.mean(d) for d in data]

# Data for secondary function
x_line = np.linspace(1, 5, 100) 
y_line = np.sin(x_line)

fig, ax1 = plt.subplots()

# Boxplot on primary y-axis
ax1.plot(x, mean_values, marker='o', label='Mean of each group')
ax1.set_xlabel('X')
ax1.set_ylabel('Distribution and Mean')

# Create secondary y-axis
ax2 = ax1.twinx()
ax2.plot(x_line, y_line, linestyle='--', label='sin(x)')
ax2.set_ylabel('sin(x)')

# Combine legends from both axes
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines1 + lines2, labels1 + labels2, loc='upper left')

fig.tight_layout()
plt.show()
```

## Triple Axes Marking

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
x = np.linspace(0, 2 * np.pi, 400)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)

# Create figure and first axis
fig, ax1 = plt.subplots()

# Create second and third y-axes
ax2 = ax1.twinx()
ax3 = ax1.twinx()
ax3.spines["right"].set_position(("outward", 60))  # offset third axis

# Plot the three functions
ax1.plot(x, y1, label="sin(x)")
ax2.plot(x, y2, label="cos(x)")
ax3.plot(x, y3, label="tan(x)")

# Set axis labels
ax1.set_xlabel("x (radians)")
ax1.set_ylabel("sin(x)")
ax2.set_ylabel("cos(x)")
ax3.set_ylabel("tan(x)")

# Limit tan values for clarity
ax3.set_ylim(-10, 10)

# Apply solid y-axis grid lines on all axes
for ax in (ax1, ax2, ax3):
    ax.yaxis.grid(True, linestyle='-')

# Combine legends from all axes
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
lines3, labels3 = ax3.get_legend_handles_labels()
ax1.legend(lines1 + lines2 + lines3, labels1 + labels2 + labels3, loc="upper left")

plt.title("sin(x), cos(x), and tan(x) with Three Y-Axes (Solid Y-Grid Lines)")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
Q = np.array([4.743, 4.243, 3.354, 2.598, 2.121])
H = np.array([7.418, 11.727, 16.035, 19.684, 22.162])
Pout = np.array([370.88, 504.54, 534.03, 505.01, 461.3])
Efficiency = np.array([30.08, 39.48, 40.7, 40.34, 40.42])

# Curve fits (2nd degree polynomial)
coeffs_H = np.polyfit(Q, H, 2)
coeffs_Pout = np.polyfit(Q, Pout, 2)
coeffs_Eff = np.polyfit(Q, Efficiency, 2)

poly_H = np.poly1d(coeffs_H)
poly_Pout = np.poly1d(coeffs_Pout)
poly_Eff = np.poly1d(coeffs_Eff)

# Fine Q for smooth curves
Q_fit = np.linspace(Q.min()*0.9, Q.max()*1.1, 200)

# Create figure and axes
fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax3 = ax1.twinx()
ax3.spines["right"].set_position(("outward", 60))

# Plot data points
ax1.scatter(Q, H, color='tab:blue', marker='o', label='H')
ax2.scatter(Q, Pout, color='tab:green', marker='s', label='Pout')
ax3.scatter(Q, Efficiency, color='tab:red', marker='^', label='Efficiency')

# Plot fitted curves
ax1.plot(Q_fit, poly_H(Q_fit), color='tab:blue', linestyle='-')
ax2.plot(Q_fit, poly_Pout(Q_fit), color='tab:green', linestyle='-')
ax3.plot(Q_fit, poly_Eff(Q_fit), color='tab:red', linestyle='-')

# Labels
ax1.set_xlabel('$Q$')
ax1.set_ylabel('$H$', color='tab:blue')
ax2.set_ylabel('$P_{out}$', color='tab:green')
ax3.set_ylabel('$Efficiency$', color='tab:red')

# Legends
lines = []
labels = []
for ax in [ax1, ax2, ax3]:
    l, lbl = ax.get_legend_handles_labels()
    lines += l
    labels += lbl
ax1.legend(lines, labels, loc='best')

# plt.title('H, Pout, and Efficiency vs Q with Curve Fits')
plt.tight_layout()
plt.savefig('my_figure.png', dpi=600)
plt.show()
```
