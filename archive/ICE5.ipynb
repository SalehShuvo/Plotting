```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("ice5.csv")
df
```

```python
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.serif'] = ['DejaVu Serif']

df_sorted = df.sort_values('Electric Power Output')
x = df_sorted['Electric Power Output']
y = df_sorted['Exhaust Gas Temp.']

fig, ax1 = plt.subplots()


line1, = ax1.plot(
    x, y,
    color='purple', marker='o',
    label='Exhaust Gas Temp.', linestyle='-'
)

ax1.set_xlabel('Electric Power Output (kW)')
ax1.set_ylabel('Exhaust Gas Temperature  (C)')






# lines = [line1]
# labels = [line.get_label() for line in lines]
# fig.legend(lines, labels, loc='upper center', bbox_to_anchor=(0.5, 1.1), ncol=2)




plt.title(
    'Fig.1 : Variation Of Exhaust Gas Temperature With Electric Power Output',
    y=-0.2
)
fig.subplots_adjust(bottom=0.2)
fig.tight_layout()
plt.grid(True)
plt.savefig('ICE5_1', dpi=300, bbox_inches='tight')
plt.show()

```

```python
df_sorted = df.sort_values('Bkw')
x = df_sorted['Bkw']
y = df_sorted['Bsfc']

fig, ax1 = plt.subplots()


line1, = ax1.plot(
    x, y,
    color='purple', marker='o',
    label='Exhaust Gas Temp.', linestyle='-'
)

ax1.set_xlabel('Bkw (kW)')
ax1.set_ylabel('Bsfc (Nm$^3$/bkw.hr)')







# lines = [line1]
# labels = [line.get_label() for line in lines]
# fig.legend(lines, labels, loc='upper center', bbox_to_anchor=(0.5, 1.1), ncol=2)




plt.title(
    'Fig.2 : Variation Of Bsfc With Bkw',
    y=-0.2
)
fig.subplots_adjust(bottom=0.2)
fig.tight_layout()
plt.grid(True)
plt.savefig('ICE5_2', dpi=300, bbox_inches='tight')
plt.show()
```
